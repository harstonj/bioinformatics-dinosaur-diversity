## Technique:

### Create a pandas DataFrame
The raw fossil data was dowloaded as a .csv file, which I first had to format so pandas could use it by removing the header info. I then saved the data as a pandas DataFrame. 

```python
import pandas as pd
from pandas import read_csv

#load the text version of the table (a csv file) into python using pandas
feature_table = read_csv("vignette.csv")

feature_table
```

### Sort the data
To obtain the data needed for my problem, I needed to sort the data by state and species type. I used a pandas feature called crosstab that returns a DataFrame that is a cross tabulation of two or more factors. This computed a frequency table of the species in each state.

```python
states = feature_table.loc[:,'state']
species = feature_table.loc[:,'name']
    
table1 = pd.crosstab(states,species)
table1
```

### Calculate alpha diversity
With the data sorted in a new table, I used a dictionary to store the states and their alpha diversity counts. Using for loops, I went through each state, started the diversity count at 0, then looped through each species. If there was a species count present, I increased the alpha count for that state by 1.

```python
#calculate alpha diversity using for loops
adiv_by_state = {}
for state in states:
    adiv = 0
    state_otus = table1.loc[state,:]
    for otu_count in state_otus:
        if otu_count > 0:
            adiv += 1
    adiv_by_state[state] = adiv
    
print("Alpha-Diversity by State:")
print(adiv_by_state)
```
